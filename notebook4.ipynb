**Creating a List**

```python
# Empty
print([])
# 1D
print([1,2,3,4,5])
# 2D
print([1,2,3,[4,5]])
# 3D
print([1,2,3,[4,5,[6,7],[8,9]]])
# Hetrogenous
print([1,True,5.6,5+6j,'hello'])
# Using type conversion
print(list('hello'))
```

**Accessing Itms from a list**

```python
#Indexing
L = [1,2,3,4,5]
# positive
print(L[0])
#negative
print(L[-1])
#slicing
L = [1,2,3,4,5,6]
print(L[::-1])
```

**Adding items to a list**

```python
# append
L = [1,2,3,4,5]
L.append(True)
print(L)

L = [1,2,3,4,5]
L.append([1,2,3])
print(L)
```

```python
# extend
L = [1,2,3,4,5]
L.extend([6,7,8])
print(L)

L.extend('delhi')
print(L)
```

```python
#insert
L = [1,2,3,4,5]

L.insert(1,100)
print(L)
```

**Editing items in a list**

```python
# string are immutable
s = 'hello'
s[0] = 'x'
print(s)
```

```python
# list are mututable
L = [1,2,3,4,5]
L[-1] = 500

L[1:4] = [200,300,400]
print(L)
```

**Deleting items from a list**

```python
#del
L = [1,2,3,4,5]
# print(L)

# del L
# print(L)


#indexing
L = [1,2,3,4,5]
del L[-1]

#slicing
del L[1:3]
print(L)
```

```python
# remove -> del the value
L = [1,2,3,4,5]

L.remove(2)

print(L)
```

```python
#pop
L = [1,2,3,4,5]

L.pop()

print(L)
```

```python
#clear
L = [1,2,3,4,5]

L.clear()

print(L)
```

**Operations on list**

```python
# arithmetic
L1 = [1,2,3,4]
L2 = [5,6,7,8]

# concatenation | merge
print(L1 + L2)
```

```python
print(L1*3)
```

```python
# membership
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]

print(5 not in L1)
print(5 in L2)
print([5,6] in L2)
```

```python
#Loops
L1 = [1,2,3,4,5]
L2 = [1,2,3,4,[5,6]]

for i in L2:
  print(i)
```

**List Functions**

```python
#len|min|max|sorted
L = [2,1,3,4,7]

print(len(L))
print(min(L))
print(max(L))
print(sorted(L,reverse=True))
print(sorted(L))
```

```python
#count
L = [1,2,3,4,5,7]
L.count(5)
```

```python
#index
L = [1,2,3,4,1,4,5]
L.index(1)
```

```python
#reverse
L = [1,2,3,4,7,0]
# permanently reverse
L.reverse()
print(L)
```

```python
#sort (vs sorted)
L = [2,1,5,7,0]
print(L)
print(sorted(L))
print(L)
# permanentaly sort
L.sort()
print(L)
```

```python
# copy -> shallow
L = [2,1,5,7,0]
print(L)
print(id(L))
L1 = L.copy()
print(L1)
print(id(L1))
```
