**Variable**

*   Static vs Dynamic typing
*   Static vs Dynamic binding
*   stylish declaration techniques





```python
name = 'nitish'
print(name)

a = 5
b = 6
print(a+b)
```

```python
#Dynamic typing
a = 5 # python support dynamic typing
# Static typing
# int a = 5 -> in other languages
```

```python
#Dynamic binding
a = 5 # python support dynamic binding
print(a)
a = 'nitish'
print(a)

# Static binding
# int a = 5; -> in other languages
# int a = "nitish" -> its not possible
```

```python
a = 5
b = 6
c = 7
print(a,b,c)
```

```python
a,b,c = 5,6,7
print(a,b,c)
```

```python
a=b=c= 5
print(a,b,c)
```
