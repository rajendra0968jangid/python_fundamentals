```python
#Pascal case
# HelloWorld
class Atm:

  #constructor
  def __init__(self):
    self.pin = ''
    self.balance = 0
    print("constructor execute")

  def menu(self):
    user_input = input("""
    Hi how can i help you !
    1. Press 1 to create pin
    2. Press 2 to change pin
    3. Press 3 to check balance
    4. Press 4 to withdraw
    5. Exit
    """)

    #conditions
    if user_input == "1":
      # create pin
      self.create_pin()
    elif user_input == "2":
      # change pin
      self.change_pin()
    elif user_input == "3":
      # check balance
      self.check_balance()
    elif user_input == "4":
      # withdraw
      self.withdraw()
    else:
      # exit
      exit()

  def create_pin(self):
    user_pin = input('Enter your pin:')
    self.pin = user_pin

    user_balance = int(input('Enter your balance:'))
    self.balance = user_balance

    print('Pin created successfully !')

  def change_pin(self):
    user_pin = input('Enter your pin:')
    self.pin = user_pin

    print('Pin updated successfully !')

  def check_balance(self):
    print("Your balance is:",self.balance)

  def withdraw(self):
    user_balance = int(input('Enter withdraw balance:'))
    self.balance = user_balance
```
