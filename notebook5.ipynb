**Tuples**
A tuple in python is similar to list.

The difference between the two is that we cannot change the elements of a tuple once it is assigned where as we can change the element of a list.

In short, a tuple is an immutuable list.

A tuple can not be changed in any way once it is created.

**Characterstics**


*   Ordered
*   Unchanged
*   Allows duplicate



**Plan of Attack**


*   Creating a Tuple
*   Accessing items
*   Editing items
*   Adding items
*   Deleting items
*   Operations on tuples
*   Tuple functions



**Creating tuple**

```python
# empty
t1 = ()
print(t1)

# creating a tuple with a single item
t2 = (2)
print(t2)
print(type(t2))

t2 = ('hello')
print(t2)
print(type(t2))
```

```python
#homogenous tuple
t3 = (1,2,3,4)
print(t3)

# heterogenous tuple
t4 = (1,2,3,True,[1,2,3])
print(t4)

t5 = (1,2,3,(4,5))
print(t5)

#using type conversion
t6 = tuple('hello')
print(t6)
```

**Accessing Items**


*   Indexing
*   Slicing



```python
#Indexing
t3 = (1,2,3,True,(4,5))
print(t3)
print(t3[0])
print(t3[4])
print(t3[4][0])
```

```python
#Slicing
t3 = (1,2,3,True,(4,5))
print(t3)
print(t3[-1][0])
```

**Editing Items**

```python
print(t3) # not support the edit
t3[0] = 100
```

**Adding Items**

```python
print(t3)
# edit not possible
```

**Deleting items**

```python
print(t3)
del t3
print(t3)
```

```python
# remove a part from tuple | not works
print(t5)
del t5[-1]
```

**Operations on tuples**

```python
# + and *
t1 = (1,2,3,4,5)
t2 = (5,6,7,8)

print(t1 + t2)

print(t1*3)

#membership
print(1 in t1)

#iteration
for i in t1:
  print(i)
```

**Tuple functions**

```python
# len/sum/min/max/sorted
t = (1,11,3,4)
print(len(t))

print(sum(t))

print(min(t))

print(max(t))

print(sorted(t))

print(sorted(t,reverse=True))
```

```python
 #count
t = (1,2,3,4,5)

t.count(5)
```

```python
#index
t.index(3)
```

**Difference between list and tuples**


*   Syntax
*   Mutability
*   Speed
*   Memory
*   Built in functionality
*   Error prone
*   Usability



```python
import time

L = list(range(100000000))
T = list(range(100000000))

start  = time.time()

for i in L:
  i*5
print('List time:',time.time()-start)

start = time.time()

for i in T:
  i*5
print('Tuple time:',time.time()-start)
```

```python
import sys

L = list(range(1000))
T = tuple(range(1000))

print('List size:',sys.getsizeof(L))
print('Tuple size:',sys.getsizeof(T))
```

```python
# possible in list beacause mutuable
a = [1,2,3]

b = a

a.append(4)
print(a)
print(b)
```

```python
# immutuable
a = (1,2,3)
b = a

a = a + (4,)

print(a)
print(b)
```

**Special Syntax**

```python
# tuple unpacking
a,b,c = (1,2,3)
print(a,b,c)
```

```python
a = 10
b = 12
a,b = b,a

print(a,b)
```

**Sets**

A sets is an unordered collection of items.Every set elements is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union,intersection,symmetric difference etc.

Characterstics:-


*   Unordered
*   Mutable
*   No Duplicate
*   Can't contain mutable data types



**Creating Sets**

```python
#empty
s = {}
print(s)
print(type(s)) # default is dictionary

s = set()
print(s)
print(type(s))
#1D and 2D
s1 = {1,2,3}
print(s1)

# 2D
# s2 = {1,2,3,{4,5}}
# print(s2) # not possible in set

# homo and hetro
s3 = {1,'hello',4.5,True,(1,2,3)}
print(s3) # True treat as a 1 duplicate not allowed in this set


# using type conversion
s4 = set([1,2,3])
print(s4)
# duplicates not allowed
s5 = {1,1,2,3,4,4,5,6}
print(s5)
# set can't have multiple items
s6  = {1,2,[3,4]}
print(s6)
```

```python
# question
s1 = {1,2,3}
s2 = {3,2,1}
print(s1 == s2)
```

**Accessing Items**

```python
s1 = {1,2,3,4}
print(s1[0]) # you can't do index in set
print(s1[-1])
```

**Editing Items**

```python
# editing not working in set
s1 = {1,2,3,4}
s1[0] = 100
```

**Adding Items**

```python
s = {1,2,3,4}
# add
s.add(5)
print(s)
# udpate
s.update([5,6,7])
print(s)
```

**Deleting Items**

```python
#del
s = {1,2,3,4,5}
# print(s)
# del s
# print(s)

# discard
s = {1,2,3,4}
s.discard(2)
print(s)
# remove
s.remove(3)
print(s)
# pop
s.pop()
print(s)
# clear
s.clear()
print(s)
```
