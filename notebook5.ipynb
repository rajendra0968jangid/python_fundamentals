**Tuples**
A tuple in python is similar to list.

The difference between the two is that we cannot change the elements of a tuple once it is assigned where as we can change the element of a list.

In short, a tuple is an immutuable list.

A tuple can not be changed in any way once it is created.

**Characterstics**


*   Ordered
*   Unchanged
*   Allows duplicate



**Plan of Attack**


*   Creating a Tuple
*   Accessing items
*   Editing items
*   Adding items
*   Deleting items
*   Operations on tuples
*   Tuple functions



**Creating tuple**

```python
# empty
t1 = ()
print(t1)

# creating a tuple with a single item
t2 = (2)
print(t2)
print(type(t2))

t2 = ('hello')
print(t2)
print(type(t2))
```

```python
#homogenous tuple
t3 = (1,2,3,4)
print(t3)

# heterogenous tuple
t4 = (1,2,3,True,[1,2,3])
print(t4)

t5 = (1,2,3,(4,5))
print(t5)

#using type conversion
t6 = tuple('hello')
print(t6)
```

**Accessing Items**


*   Indexing
*   Slicing



```python
#Indexing
t3 = (1,2,3,True,(4,5))
print(t3)
print(t3[0])
print(t3[4])
print(t3[4][0])
```

```python
#Slicing
t3 = (1,2,3,True,(4,5))
print(t3)
print(t3[-1][0])
```

**Editing Items**

```python
print(t3) # not support the edit
t3[0] = 100
```

**Adding Items**

```python
print(t3)
# edit not possible
```

**Deleting items**

```python
print(t3)
del t3
print(t3)
```

```python
# remove a part from tuple | not works
print(t5)
del t5[-1]
```

**Operations on tuples**

```python
# + and *
t1 = (1,2,3,4,5)
t2 = (5,6,7,8)

print(t1 + t2)

print(t1*3)

#membership
print(1 in t1)

#iteration
for i in t1:
  print(i)
```

**Tuple functions**

```python
# len/sum/min/max/sorted
t = (1,11,3,4)
print(len(t))

print(sum(t))

print(min(t))

print(max(t))

print(sorted(t))

print(sorted(t,reverse=True))
```

```python
 #count
t = (1,2,3,4,5)

t.count(5)
```

```python
#index
t.index(3)
```

**Difference between list and tuples**


*   Syntax
*   Mutability
*   Speed
*   Memory
*   Built in functionality
*   Error prone
*   Usability



```python
import time

L = list(range(100000000))
T = list(range(100000000))

start  = time.time()

for i in L:
  i*5
print('List time:',time.time()-start)

start = time.time()

for i in T:
  i*5
print('Tuple time:',time.time()-start)
```

```python
import sys

L = list(range(1000))
T = tuple(range(1000))

print('List size:',sys.getsizeof(L))
print('Tuple size:',sys.getsizeof(T))
```

```python
# possible in list beacause mutuable
a = [1,2,3]

b = a

a.append(4)
print(a)
print(b)
```

```python
# immutuable
a = (1,2,3)
b = a

a = a + (4,)

print(a)
print(b)
```
