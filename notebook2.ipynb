```python
#sequence sum
# 1/1 + 2/2 + 3/3 + .....
#code here

n = int(input('enter n:'))

result = 0
fact = 1

for i in range(1,n+1):
  fact = fact * i
  result = result + i/fact

print(result)
```

**Nested Loop**

```python
#Example -> unique pairs
for i in range(1,5):
  for j in range(1,5):
    print(i,j)
```

```python
#Patterns
#*
#**
#***

#code here

row = int(input('Enter number of rows:'))

for i in range(1,row+1):
  for j in range(1,i+1):
    print("*",end="  ")
  print("\n")
```

**Loop Control Statement**

*   break
*   continue
*   pass



```python
for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
lower = int(input('enter lower range:'))
upper = int(input('enter upper range:'))

for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j == 0:
      break
    else:
      print(i)
```

```python
#Continue
for i in range(1,10):
  if i == 5:
    continue
  print(i)
```

```python
#pass
for i in range(1,5):
  pass # save you from error
```

```python
#Accessing sub-string from string
#Positive Indexing
s = 'hello World'
print(s[6])
```

**Editing and Deleting in strings**

```python
s = 'hello world'
s[0] = 'H'
# Python string are immutable
```

```python
#Deleting strings
s = 'hello world'
del s
print(s)
```

```python
# Delete the string but can't delete the sub part of string
s = 'hello world'
del s[-1:-5:2]
print(s)
```
