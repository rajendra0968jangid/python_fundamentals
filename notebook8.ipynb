```python
# required argument
def si(p,t,r):
  print((p*r*t)/100)
si(100,2,5)
```

```python
# default argument
def si(p,t,r=10.5):
  print((p*r*t)/100)
si(100,2)
```

```python
# keyword argument
def si(p,t,r=10.5):
  print((p*r*t)/100)
si(100,2,8.5)
```

```python
# variable argument
def sample(*a):
  print(a)
sample(1,2,3,4,5)
```

```python
# task 1
def sum(*a):
  total = 0
  for i in a:
    total += i
  print("sum:-",total)

sum(1,2,3,4,5)
```

```python
# task 2
def sum(*a):
  total = 0
  for i in a:
    total += i
  print("average:-",total/len(a))

sum(1,2,3,4,5)
```

```python
# maximum
def mymax(*a):
  m = a[0]
  for i in a:
    if(m < i):
      m = i
  print(m)
mymax(1,5,7,8,9,12,2,3)
```

```python
class rectangle:
  def __init__(self,l,b):
    self.l = l
    self.b = b
  def area(self):
    print("Area of rectangle is :", self.l*self.b)
  def parameter(self):
    print("parameter of rectangle: ", 2 * (self.l+self.b))
obj = rectangle(10,20)
obj.area()
obj.parameter()
```

```python
# function over riding
class parent:
  def bmw(self):
    print("bmw of parents class :")
  def honda(self):
    print("honda of parent class :")

class child(parent):
  def bmw(self):
    super().bmw()
    print("bmw of child class :")
  def honda(self):
    print("honda of child class :")
objp = parent()
objp.bmw()

objc = child()
objc.bmw()
```

```python
class gp:
  def name(self):
    print("gp name :")
class parent(gp):
  def name(self):
    super().name()
    print("parent name :")
class child(parent):
  def name(self):
    super().name()
    print("child name :")

objc = child()
objc.name()
```
