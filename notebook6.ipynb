**Lets create a function** (with docstring)

```python
def is_even(num):
  """
  This function returns if a given number is odd or even
  input - any valid integer
  output - even/odd
  """
  if type(num) != int:
    return "Not a valid integer"
  if num % 2 == 0:
    return 'even'
  else:
    return 'odd'

#function call
for i in range(1,11):
  x = is_even(i)
  print(i,x)
```

```python
# default argumen
def power(a=1,b=1):
  return a**b
```
